```python
import pandas as pd
```

```python
df=pd.read_csv("data/Telco-Customer-Churn.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df[df["TotalCharges"].isnull()][["tenure","MonthlyCharges","TotalCharges"]]
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
df["Churn"]=df["Churn"].map({"Yes":1,"No":0})
df["Churn"].value_counts()
```

```python
churn_rate=df["Churn"].mean()*100
churn_rate
```

```python
pd.crosstab(df["Contract"],df["Churn"],normalize="index")*100
```

```python
df.groupby("Churn")["tenure"].mean()
```

```python
df["tenure_group"]=pd.cut(df["tenure"],bins=[0,12,24,48,72],labels=["0-1 Years","1-2 Years","2-4 Years","4-6 Years"])
pd.crosstab(df["tenure_group"],df["Churn"],normalize="index")*100
```

```python
df.groupby("Churn")["MonthlyCharges"].mean()
```

```python
pd.crosstab(df["InternetService"],df["Churn"],normalize="index")*100
```

```python
pd.crosstab(df["PaymentMethod"],df["Churn"],normalize="index")*100
```

```python
pd.crosstab([df["Contract"],df["PaymentMethod"]],df["Churn"],normalize="index")*100
```

```python
df[df["Churn"]==1]["MonthlyCharges"].sum()
```
